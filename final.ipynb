| Step                             | Tool/Tech                                                                       | What we do                                                                               |
| -------------------------------- | ------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- |
| **1) NLP Preprocessing**         | `spaCy` or `NLTK`                                                               | Tokenize, POS tag, Lemmatize module descriptions and questions                           |
| **2) Text Vectorization**        | Bag of Words or TF-IDF (`sklearn`)                                              | Convert text to numeric vectors                                                          |
| **3) Task Categorization Model** | Logistic Regression, Random Forest                                              | Train model to predict e.g. course level (Beginner, Intermediate, Advanced) or task type |
| **4) Time-Series Forecasting**   | Prophet or ARIMA                                                                | Predict trends like number of modules consumed over time                                 |
| **5) Smart Task Prediction**     | Hybrid of ML + Rules                                                            | E.g., If “supervised learning” in question → boost supervised modules                    |
| **6) Evaluate**                  | Accuracy for classifier, RMSE for time-series, Precision/Recall for multi-class |                                                                                          |


```python
import pandas as pd
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from prophet import Prophet
```

```python
target_col ='Course_Level'
```

```python
#NLP processing with spacy
nlp = spacy.load('en_core_web_sm')
def preprocess(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return " ".join(tokens)

modules_df['processed_text'] = modules_df['Module'].astype(str).apply(preprocess)
```

```python
#vectorisation
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(modules_df['processed_text'])
y = modules_df[target_col]
```

```python
#train classifier
#logistic regression
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
#evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
```

```python
#random forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print("Random Forest Accuracy:", rf.score(X_test, y_test))
```
